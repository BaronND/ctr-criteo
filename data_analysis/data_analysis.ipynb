```python
categorical_feature = [var for var in data.columns if 'C' in var]
contionous_feature = [var for var in data.columns if 'I' in var]
```

```python
print(categorical_feature)
```

```python
print(contionous_feature)
```

```python
# 数值变量之间的相关性
correlation_data = data[contionous_feature].corr()
```

```python
print(correlation_data)
```

```python
# 打印相关性高的变量
threshold = 0.5
for i in range(len(correlation_data)):
    for j in range(len(correlation_data)):
        if(i > j) and (correlation_data.iloc[i, j] > threshold):
            print('%s and %s = %.2f' % (data.columns[i+1], data.columns[j+1], correlation_data.iloc[i, j]))
```

```python
# 画图查看分类变量的分布
for feature in categorical_feature:
    sns.countplot(x=data[feature], data=data)
    plt.show()
```
